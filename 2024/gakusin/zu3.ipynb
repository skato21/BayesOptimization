```python
import numpy as np
import matplotlib.pyplot as plt

# Values and errors for A and D
data = {
    'babar': {'A_value': -0.25, 'A_error': 3.05, 'D_value': 6.65, 'D_error': 4.95},
    'belle': {'A_value': -0.4, 'A_error': np.sqrt((1.4)**2 + (0.3)**2), 'D_value': 6.2, 'D_error': np.sqrt((1.5)**2 + (0.6)**2 + (1.2)**2)},
    'belle2': {'A_value': -0.3, 'A_error': np.sqrt((0.44)**2 + (0.24)**2), 'D_value': 6.3, 'D_error': np.sqrt((0.47)**2 + (0.48)**2 + (0.96)**2)},
    'belle3': {'A_value': -0.3, 'A_error': np.sqrt((0.16)**2 + (0.24)**2), 'D_value': 6.3, 'D_error': np.sqrt((0.18)**2 + (0.48)**2 + (0.96)**2)}
}

# Extracting data
experiments = list(data.keys())
A_values = [data[exp]['A_value'] for exp in experiments]
A_errors = [data[exp]['A_error'] for exp in experiments]
D_values = [data[exp]['D_value'] for exp in experiments]
D_errors = [data[exp]['D_error'] for exp in experiments]

# Plotting value +- error for A and D
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# A values plot
axs[0].errorbar(experiments, A_values, yerr=A_errors, fmt='o', capsize=5, color='black', label='A')
axs[0].set_ylabel('$A_{cp}$')
axs[0].set_title('A Value ± Error')
axs[0].legend()
#axs[0].grid(True)

# D values plot
axs[1].errorbar(experiments, D_values, yerr=D_errors, fmt='o', capsize=5, color='black', label='D')
axs[1].set_ylabel('$\Delta_{cp}$')
axs[1].set_title('D Value ± Error')
axs[1].legend()
#axs[1].grid(True)

# Common X axis
for ax in axs:
    ax.set_xticks(np.arange(len(experiments)))
    ax.set_xticklabels(experiments)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Values and errors for A and D
data = {
    'babar': {'A_value': -0.25, 'A_error': 3.05, 'D_value': 6.65, 'D_error': 4.95},
    'belle': {'A_value': -0.4, 'A_error': np.sqrt((1.4)**2 + (0.3)**2), 'D_value': 6.2, 'D_error': np.sqrt((1.5)**2 + (0.6)**2 + (1.2)**2)},
    'belle2': {'A_value': -0.4, 'A_error': np.sqrt((0.44)**2 + (0.24)**2), 'D_value': 6.3, 'D_error': np.sqrt((0.47)**2 + (0.48)**2 + (0.96)**2)},
    'belle3': {'A_value': -0.4, 'A_error': np.sqrt((0.16)**2 + (0.12)**2), 'D_value': 6.3, 'D_error': np.sqrt((0.18)**2 + (0.24)**2 + (0.48)**2)}
}

# Extracting data
experiments = list(data.keys())
A_values = [data[exp]['A_value'] for exp in experiments]
A_errors = [data[exp]['A_error'] for exp in experiments]
D_values = [data[exp]['D_value'] for exp in experiments]
D_errors = [data[exp]['D_error'] for exp in experiments]

# Plotting value +- error for A and D

plt.rcParams["font.size"] = 25
fig, axs = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# A values plot
A_errorbar = axs[0].errorbar(experiments, A_values, yerr=A_errors, fmt='o', capsize=5, color='black', label='Measurement',elinewidth=2, capthick=2)
axs[0].set_ylabel('$A_{cp}$ $[×10^{-2}]$ ',fontsize = 30)
axs[0].grid(True)
sm_patch_1 = axs[0].axhspan(-0.2, 0.2, edgecolor='black', facecolor='none', alpha=0.5, hatch='//', label='SM prediction[1]')
axs[0].legend(handles=[sm_patch_1, A_errorbar], labels=['SM prediction[1]', 'Measurement'])
axs[0].set_yticks(np.linspace(-3, 3, 7))  # X軸の目盛りを0から10まで11個配置
axs[0].tick_params(axis='both', which='major', labelsize=30)

# D values plot
D_errorbar = axs[1].errorbar(experiments, D_values, yerr=D_errors, fmt='o', capsize=5, color='black', label='Measurement',elinewidth=2, capthick=2)
axs[1].set_ylabel('$\Delta_{0+}$ $[×10^{-2}]$', fontsize = 30)
axs[1].grid(True)
sm_patch_2 = axs[1].axhspan(2.3, 7.5, edgecolor='black', facecolor='none', alpha=0.5, hatch='//', label='SM prediction[1]')
axs[1].legend(handles=[sm_patch_2, D_errorbar], labels=['SM prediction[1]', 'Measurement'])
axs[1].set_ylim(bottom=0)
axs[1].set_yticks(np.linspace(0, 12, 7))  # X軸の目盛りを0から10まで11個配置
axs[1].tick_params(axis='both', which='major', labelsize=30)

# Common X axis
axs[1].set_xticks(np.arange(len(experiments)))
axs[1].set_xticklabels([])
#axs[1].set_xticklabels(experiments)
#axs[1].set_xlabel('Experiments')

plt.tight_layout()
plt.show()
```
