```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_fontja
# 電流の初期値と最終値 (mA)
initial_current = 1000  # mA
final_current = 800  # mA

# 入射効率の設定
efficiency_start = 0.40  # 40% の入射効率
efficiency_end = 0.20  # 20% の入射効率

# 減衰にかかる時間範囲の設定 (任意単位)
time = np.linspace(0, 10, 500)  # 0から10までの時間範囲を500分割

# 減衰定数の計算
decay_constant = -np.log(final_current / initial_current)  # 電流の変化に基づく減衰定数

# 指数関数的な減衰の計算式
current = initial_current * np.exp(-decay_constant * time)

# グラフの描画
plt.figure(figsize=(8, 6))
plt.plot(time, current, label="電流減衰", color='b')

# グラフのタイトルとラベル
plt.title("入射効率の低下による電流の減衰")
plt.xlabel("時間 (任意単位)")
plt.ylabel("電流 (mA)")
plt.grid(True)

# 終点の電流値を強調するために横線を描画（定電流ライン）
plt.axhline(y=final_current, color='r', linestyle='--', label=f"最終電流: {final_current} mA")
plt.axhline(y=initial_current, color='g', linestyle='--', label=f"初期電流: {initial_current} mA")

plt.legend()

# グラフの表示
plt.show()
```
