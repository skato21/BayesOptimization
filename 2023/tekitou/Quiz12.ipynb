```python
import numpy as np
import matplotlib.pyplot as plt

# 逆関数法による指数分布生成
def generate_exponential(tau, size):
    u = np.random.uniform(0, 1, size)
    x = -tau*np.log(1 - u)
    return x

# パラメータ設定
tau = 2  # τの値
sample_size = 100  # 生成するデータの数

# 指数分布生成
data = generate_exponential(tau, sample_size)

# ヒストグラム表示
plt.hist(data, bins=30, density=True, alpha=0.7, label='Generated Data')

# 指数関数を描画
x = np.linspace(0, 10, 1000)
y = (1/tau) * np.exp(-x/tau)
plt.plot(x, y, 'r', label='ideal exponential function')

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```
